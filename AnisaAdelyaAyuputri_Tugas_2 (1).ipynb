# Penjelasan Dataset



Dataset yang digunakan bersumber dari kaggle yaitu berjudul Psycological Effects of COVID. Dataset ini merupakan jawaban survei yang diambil pada tahun 2020 untuk memahami pendapat orang tentang lockdown. Dataset ini dikumpulkan dengan bantuan Google Formulir. Dataset berisi jawaban atas berbagai pertanyaan yang diberikan oleh orang-orang. Isi kolom dari dataset tersebut adalah :

* age : Age group of the person
* gender : Gender of the person
* occupation : Occupation/sector where the person works
* line_of_work : The line of work performed by the person
* time_bp : The time spent on work before pandemic
* time_dp : The time spent on work during pandemic
* travel_time : The travel time spent
* easeof_online : Rating of work going online
* home_env : Liking of home environment
* prod_inc " Rating Productivity Increase
* sleep_bal : The rating of sleep cycle
* new_skill : Whether any new skill was learnt
* fam_connect : Rating how well the person connected with his family
* relaxed : Rating of how relaxed the person is feeling
* self_time : Rating how much self time was procured
* like_hw : Liking of working from Home
* dislike_hw : Disliking Working from Home
* prefer : Preference of the person to work from home/office
* certaindays_hw : Liking whether certain days of working from home is needed




# Visualisasi Data

```python
import pandas as pd
# Pandas adalah library Python yang digunakan untuk manipulasi dan analisis data.
# Pandas ini digunakan untuk membaca dan menulis berbagai jenis file data,
# menggabungkan dan mengubah data, serta melakukan operasi statistik pada data.
import seaborn as sns
# Seaborn adalah library visualisasi data berdasarkan Matplotlib.
# Seaborn memiliki banyak fungsi untuk membuat plot seperti scatter plot, line plot, bar plot, histogram, heatmap, dan banyak lagi.
import matplotlib.pyplot as plt
# Pyplot adalah modul dalam Matplotlib yang digunakan untuk membuat berbagai jenis plot,
# seperti bar plot, line plot, scatter plot, histogram, dan sebagainya.
import numpy as np
# NumPy adalah library dasar untuk komputasi numerik di Python.
```

```python
# Memanggil dataset yang ada sudah di upload di github
dataset = "https://raw.githubusercontent.com/anisa-adelya/tugas-ds-2/main/psyco.csv"
df = pd.read_csv(dataset, encoding='unicode_escape')
# Menampilkan 5 data teratas pada dataset
df.head()
```

```python
# Membuat barplot menggunakan modul matplotlib.pyplot
# Contoh yang dilakukan adalah melihat jumlah partisipan berdasarkan jenis kelamin
gender_counts = df['gender'].value_counts()
gender_counts.plot(kind='bar', title='Jumlah Partisipan berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
plt.show()
```

Berdasarkan Barplot tersebut dapat diketahui bahwa jumlah partisipan yang ikut mengisi survei ini lebih banyak lelaki dibandingkan dengan perempuan. Selain itu juga terdapat beberapa partisipan yang memilih untuk tidak menyebutkan jenis kelamin nya pada survei yang dilakukan.

```python
# Membuat dataset
df_pivot = pd.pivot_table(df, values="time_dp",
                          index="occupation",
                          columns="prefer",
                          aggfunc= np.sum)
# Membuat stacked bar chart
ax = df_pivot.plot(kind="bar",alpha=0.5, stacked = True)
# Menambah penjelasan pada stacked bar
plt.title('Total time during pandemic user by Occupation & Prefer user using Matplotlib \n')
plt.xlabel('Occupation')
plt.ylabel('Total Time During Pandemic')
# Show the plot
plt.show()
```

Dari gambar diatas kita mendapatkan stacked bar chart yang memvisualisasikan total waktu yang dihabiskan oleh partisipan selama pandemi, dengan tiap pekerjaan di sumbu X, dan total waktu di sumbu Y. Setiap bar pada plot akan terdiri dari beberapa bagian yang mewakili preferensi pengguna yang berbeda, dan jumlah totalnya akan mewakili total waktu dihabiskan oleh pengguna berdasarkan pekerjaan dan preferensi. Berdasarkan Stacked Bar Chart diatas diketahui bahwa partisipan pada survei kali ini paling banyak memiliki jumlah waktu yang dihabiskan untuk bekerja selama masa pandemi adalah yang bekerja sebagai Working Proffesional dan yang paling sedikit adalah Retired / Senior Citizen. Selain itu juga dapat diketahui bahwa partisipan lebih banyak yang memilih untuk menghadiri kegiatan nya kekantor atau kesekolah langsung daripada bekerja atau belajar secara online.

```python
# Membuat bar plot
sns.barplot(x = 'occupation',y = 'time_dp',data = df,palette = sns.color_palette('coolwarm', n_colors=5), dodge=True)

# Menambah aestetik
plt.title('Jumlah waktu yang dihabiskan selama pandemic berdasarkan kategori pekerjaannya')
plt.xlabel('Occupation')
plt.ylabel('Total Time During Pandemic')

# Mengatur rotasi label pada sumbu X
plt.xticks(rotation=90)  # Ubah angka rotasi sesuai kebutuhan

# Show plotnya
plt.show()
```

Berdasaarkan hasil pada visualisasi diatas dapat diketahui bahwa Pekerjaan yang Menghabiskan Waktu Lebih Banyak selama masa pandemik adalah pekerjaan Retired/Senior Citizen. Dilihat dari tinggi bar pada sumbu Y, kita dapat melihat jenis pekerjaan yang cenderung menghabiskan lebih banyak waktu selama pandemi. Pekerjaan dengan bar yang lebih tinggi menunjukkan bahwa individu dalam pekerjaan tersebut menghabiskan waktu lebih banyak. Selain itu dapat diketahui ada perbedaan pekerjaan dalam pengelompokan. Berdasarkan sumbu X (pekerjaan), kita dapat melihat perbedaan dalam durasi waktu yang dihabiskan antara berbagai jenis pekerjaan. Grafik akan membantu mengidentifikasi apakah ada pekerjaan tertentu yang mengalami peningkatan signifikan dalam waktu yang dihabiskan selama pandemi.

```python
# Menghitung korelasi antara kolom-kolom numerik dalam dataset
correlation_matrix = df.corr()

# Membuat heatmap dari matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title('Heatmap Korelasi antara Kolom Numerik')
plt.show()
```

Heatmap dibuat untuk mengetahui korelasi (hubungan) setiap kolom yang ada pada dataset. Korelasi kuat yaitu bernilai -1 atau 1. Nilai korelasi yang semakin mendekati nilai -1 atau 1 berarti korelasi antar kolom tersebut juga semakin kuat. Korelasi -1 menunjukkan hubungan linier sempurna yang negatif antara dua variabel. Artinya, jika salah satu variabel meningkat, variabel lain akan selalu turun dengan tingkat yang tepat. Ini sering disebut korelasi negatif sempurna. Korelasi 1 menunjukkan hubungan linier sempurna yang positif antara dua variabel. Artinya, jika salah satu variabel meningkat, variabel lain juga akan selalu naik dengan tingkat yang tepat. Ini sering disebut korelasi positif sempurna. Contoh korelasi positif yang dapat dilihat dari dataset yang digunakan yaitu untuk kolom easeof_online dengan kolom home_env yaitu bernilai 0.51, korelasi ini cukup tinggi dan hal ini menunjukkan bahwa jika nilai pengguna menyukai lingkungan rumahnya semakin tinggi maka akan semakin tinggi juga nilai rating yang diberikan untuk pekerjaan yang dilakukan secara online. Contoh lainnya korelasi yang bernilai negataif yaitu kolom prod_inc dengan kolom easeof_online yaitu bernilai -0.64 hal ini cukup tinggi karena mendekati nilai -1. Berdasarkan nilai tersebut menunjukkan bahwa semakin tinggi nilai Rating Produktivitas Meningkat maka akan semakin menurun nilai pengguna menyukai lingkungan rumahnya begitupun sebaliknya.

```python
# Menghitung distribusi umur
gender_counts = df['age'].value_counts()

# Membuat pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Distribusi Umur')
plt.axis('equal')  # Membuat lingkaran menjadi lingkaran sempurna
plt.show()
```

Berdasarkan pie chart diatas menunjukkan bahwa usia partisipan yang mengisi survey ini paling banyak berada pada rentang umur 19 - 25. Sedangkan usia partisipan yang mengisi survey ini paling sedikit berada pada rentang umur lebih dari 60.
